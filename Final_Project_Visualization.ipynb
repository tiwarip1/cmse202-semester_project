### Carnot Cycle

```python
# Initialize ivisual canvas object for visualization
scene = canvas(title='Ideal Gas')

# Initialize carnot cycle model object
cn = carnot()

# Initialize box object
boxsize = cn.boxsize
b = box(pos=((boxsize[1]+boxsize[0])/2.,
                 (boxsize[3]+boxsize[2])/2.,
                 (boxsize[5]+boxsize[4])/2.),
            size=(boxsize[1]-boxsize[0],
                  boxsize[3]-boxsize[2],
                  boxsize[5]-boxsize[4]),
            color=color.white,
            opacity=0.3)

# Set number of iterations and save inital volume
iteration = 500
print("Isothermal expansion")
initial_volume = cn.volume
label=0

# Isothermal Expansion
for i in range(iteration):
    rate(100)
    cn.update_isothermal(rate=10)   # update model, positive rate, isothermal
    
    # Update box size
    b.size=(cn.boxsize[1]-cn.boxsize[0],
                  cn.boxsize[3]-cn.boxsize[2],
                  cn.boxsize[5]-cn.boxsize[4])
    
    # Save images for presentation
#     im = ImageGrab.grab((200,200,2500,2500))
#     fname = './carnot/test' + str(label) + '.png'
#     im.save(fname)
    label += 1
    
print("Isentropic expansion")
for i in range(iteration):
    rate(100)
    cn.update_isentropic(rate=10)   # update model, positive rate, isentropic
    
    # Update box size
    b.size=(cn.boxsize[1]-cn.boxsize[0],
                  cn.boxsize[3]-cn.boxsize[2],
                  cn.boxsize[5]-cn.boxsize[4])
    
    # Save images for presentation
#     im = ImageGrab.grab((200,200,2500,2500))
#     fname = './carnot/test' + str(label) + '.png'
#     im.save(fname)
    label += 1
    
print("Isothermal compression")
for i in range(iteration):
    rate(100)
    cn.update_isothermal(rate=-10)  # update model, negative rate, isothermal
    
    # Update box size
    b.size=(cn.boxsize[1]-cn.boxsize[0],
                  cn.boxsize[3]-cn.boxsize[2],
                  cn.boxsize[5]-cn.boxsize[4])
    
    # Save images for presentation
#     im = ImageGrab.grab((200,200,2500,2500))
#     fname = './carnot/test' + str(label) + '.png'
#     im.save(fname)
    label += 1
    
print("Isentropic compression")
while cn.volume != initial_volume:
    rate(100)
    cn.update_isentropic(rate=-10)  # update model, negative rate, isentropic
    
    # Update box size
    b.size=(cn.boxsize[1]-cn.boxsize[0],
                  cn.boxsize[3]-cn.boxsize[2],
                  cn.boxsize[5]-cn.boxsize[4])
    
    # Save images for presentation
#     im = ImageGrab.grab((200,200,2500,2500))
#     fname = './carnot/test' + str(label) + '.png'
#     im.save(fname)
    label += 1
```

```python
i = len(cn.energy_arr) // 4    # Same number of iterations in each state, so this gives iterations in one state

plt.plot(cn.volume_arr[0:i], cn.pressure_arr[0:i], c='C0', label="Isothermal Expansion")
plt.plot(cn.volume_arr[i:2*i], cn.pressure_arr[i:2*i], c='C1', label="Isentropic Expansion")
plt.plot(cn.volume_arr[2*i:3*i], cn.pressure_arr[2*i:3*i], c='C2', label="Isothermal Contraction")
plt.plot(cn.volume_arr[3*i:], cn.pressure_arr[3*i:], c='C3', label="Isentropic Contraction")

plt.title("Carnot Cycle: P vs V")
plt.xlabel("Volume")
plt.ylabel("Pressure")
plt.grid()
plt.legend()

plt.savefig("./carnot/cycle.png")
```

```python
# Create 4 plots for each variable from model
fig, ax = plt.subplots(2,2)
ax[0,0].plot(np.arange(len(cn.energy_arr)), cn.energy_arr, label='Energy', c='C0')
ax[0,1].plot(np.arange(len(cn.energy_arr)), cn.temp_arr, label='Temperature', c='C1')
ax[1,0].plot(np.arange(len(cn.energy_arr)), cn.pressure_arr, label='Pressure', c='C2')
ax[1,1].plot(np.arange(len(cn.energy_arr)), cn.volume_arr, label='Volume', c='C3')

# Set x labels
ax[0,0].set_xlabel("Iterations")
ax[0,1].set_xlabel("Iterations")
ax[1,0].set_xlabel("Iterations")
ax[1,1].set_xlabel("Iterations")

# Set titles
ax[0,0].set_title("Energy")
ax[0,1].set_title("Temperature")
ax[1,0].set_title("Pressure")
ax[1,1].set_title("Volume")

# Turn on grids
ax[0,0].grid()
ax[0,1].grid()
ax[1,0].grid()
ax[1,1].grid()

# Save the figure for the presentation
fig.tight_layout()
fig.savefig("./carnot/state.png")
```
